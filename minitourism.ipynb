```python
import pandas as pd
import numpy as py
import matplotlib.pyplot as plt

```

```python
data=pd.read_csv("anil.csv",index_col="Year");
data=data[-4:]
print(data.columns)

data.columns=['Arrivals','Jan-Mar','Apr-June','July-Sep','Oct-Dec']
data.index=[0,1,2,3]
print(data[1:2][:])
```

```python

print(data)
```

```python
print(sum(data.iloc[:]['Arrivals']))
```

```python
print(data.iloc[0][1:])
```

```python
newdata=[]
for i in range(4):
    x=[data.iloc[i][0]]
    for j in data.iloc[i][1:]:
        z=data.iloc[i][0]*(j/100)
        
        x.append(z)
    newdata.append(x)
print(newdata)
```

```python

x=py.array(['2016','2017','2018','2019'])
y=[]
for i in range(4):
    y.append(newdata[i][0])
y=py.array(y)
print(x,y)
plt.bar(x,y,width=0.2,color = "#4CAF50")
plt.xlabel("Year")
plt.ylabel("Arrivals")
plt.show()
```

```python
#showing the % of distribution of arrivals over year
plt.pie(y,labels=x,autopct='%1.1f%%')
plt.show()
```

```python
for i in range(4):
    x=py.array(newdata[i][1:])
    y2=py.array(['Jan-Mar','Apr-June','July-Sep','Oct-Dec'])
    #plt.subplot(4,1,i+1)
    plt.bar(y2,x,color="#db6b6b",width=0.4)
    plt.xlabel(str(2016+i))
    plt.ylabel('Arrivals Across Year')
    plt.show()
    
```

```python
# comparision over years across each quardrants
for i in range(4):
    x=py.array(newdata[i][1:])
    y1=py.array(['Jan-Mar','Apr-June','July-Sep','Oct-Dec'])
    sizes=[200+i*100,200+i*100,200+i*100,200+i*100]
    plt.scatter(y1,x,s=sizes)
    plt.xlabel("Across Year")
    plt.ylabel('Arrivals')
    plt.legend(('2016', '2017', '2018', '2019'),scatterpoints=1,loc='lower right',ncol=3,fontsize=8)
plt.show()
    
```
